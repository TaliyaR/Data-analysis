# 3.Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?

разделим студентов на тех, у кого > 10 пропусков и тех, у кого <= 10

```python
more_ten_absence = data[data['absences'] > 10]
less_ten_absence = data[data['absences'] <= 10]
```

```python
more_ten_absence_g3 = more_ten_absence['G3'].mean()
less_ten_absence_g3 = less_ten_absence['G3'].mean()
print(more_ten_absence_g3, less_ten_absence_g3)
```

Можно сделать предположение, что студенты, имеющие больше 10 пропусков учатся немного хуже, чем остальные, но разница не велика

# 4.Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным.

на итоговую оценку могут повлиять такие признаки как: имеет дополнительные(платные) занятия, есть ли помощь семьи в выполнении задания, учавствует ли во внеклассной деятельности, состояние здоровья, употребление алкоголя(рассмотрен в примере) и тд. Практически все признаки приведенные в таблице могут прямо или косвенно влиять на итоговую оценку.

посмотрим как состояние здоровья влияет на оценку.

```python
good_health = data[data['health'] >= data['health'].mean()]
bad_health = data[data['health'] < data['health'].mean()]
```

```python
good_health_g3 = good_health['G3'].mean()
bad_health_g3 = bad_health['G3'].mean()
print(good_health_g3, bad_health_g3)
```

деалем вывод, что студенты, с плохим здоровьем имеют средний балл выше тех, кто имеет здоровье лучше

# 5.Как вы думаете, если при школе откроется общежитие, насколько актуально это будет для студентов математического класса?

```python
data_by_schoolsup = data.groupby('traveltime')
data_by_schoolsup.mean()
```

сравним средний балл студентов в зависимости от того сколько времени они тратят на дорогу до школы. Заметно, что средний балл ухудшается у тех, кто дольше добирается до школы. Поэтому общежитие при школе может способствовать обучению школьников.
